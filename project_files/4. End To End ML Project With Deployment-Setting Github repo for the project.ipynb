# 1. ***Setup Github repo for the ml project***

## - Whenever we start a ml project, since we work in a team so you need a repository to commit the code
    - setup new environment
    - decide project structure setup.py
    - requirements.text

## setup new environment
- create a new github repo : Just enter the name & Click create wihtout doing anything else



- After project created on github, move to the folder where you want to save the project
- open that path in cmd
- **Create virtual environment** using command "virtualenv venv"
- **activate virtual environment** using command "venv\Scripts\activate"
- Clone github repo & sync so that we can start commiting
- Run all those commands which we get after creating repo

- After adding readme.md file, its status will change to A which means added
- run git commit -m "first commit" 

- git push -u origin main --> pushing file from origin (my local system) to main (main branch of github)


- refresh github repo & then we'll see redme.md file which we created on local system got pushed to main branch of github repo

- **In future, if we make any change in the code or write any code, we need to push it to get updated in github repo**
- create a new gitignore file from github browser & commit
- In order to update everything from github repo to local system: **git pull** which will add gitignore file in local system



## decide project structure "setup.py"

- create 2 new files in vscode local system
- setup.py & requirements.txt

- **what is setup.py**
- It helps others install & use your project, It will build entire ml project as package which can be installed & used by anyone like pandas, numpy etc are projects which we can install & use
- Now in order setup to find packages we need "__init__.py" file so any folder have "__init__.py" file it will be treated as a package
- create a new folder src & in it create a new file "__init__.py" file
- entire project development will happen inside this project
- create setup.py which has refrence of requirements.txt file to auto install packages
- run command "pip install -r requirements.txt" in cmd
- this will create a new folder "mlproject.egg-info" which can be used to install this mlproject
- Finish it up with below commands:
- git add .
- git commit -m "updated setup.py & requirements.txt"
- git push -u origin main

- **In "SRC: folder, create other folders**
    1. "Components" folder & __init__ file so that this folder can also be created as a package, components are all the modules which are going to create like data ingestion (reading data from dbs or file location), Its part of our module of the ml project.
    2. create files "data_ingestion.py", "data_transformation.py", "model_trainer.py" file inside components folder
    3. create another folder "pipeline" & create files "__init__.py", "train_pipeline.py" & "predict_pipeline.py"
    4. since entire project implimentation will happen inside the folder "src", so lets add 3 important files:
        1. logger.py -- logging
        2. exception.py -- exception handling purpose
        3. utils.py -- for all the common helper functions        

### write code in exception handling, logger & utils python files we just created

- search "exception python documentation" in google
- Its a custom exception handling mechanism in Python. This is useful for debugging and logging errors with detailed information about where and why an error occurred.


- search logger documentation in google to understand
- Its used to log any execution happens to track them, even the errors will be logged so that we can debug them later on


- utils.py is a general functions which we will write as and when required

- run git add . to add them to be pushed
- git commit -m "created logger, exception & utils file"
- git push -u origin main
